# Array Broadcasting in Matlab Demonstration

```matlab
clear all;
close all;
% broadcasting for a simple function x+y+z, x, y are vectors given and
% choice of z minimzes the function

a_n = 100000;
z_n = 11;
k_n = 100;

%a=[1:2];
%replacement=true;
%a=randsample(50,z_n,replacement)-2.5;
a=rand(a_n,1)-0.5;
%z=[1:4];
% z=randsample(50,z_n,replacement)-2.5;
z=rand(z_n,1)-0.5;

k= linspace(-2, 2, k_n);
[a_m,z_m]=meshgrid(a,z);
a_v = a_m(:);
z_v = z_m(:);
c=0;

%func=X_v - Y_v + z + c;
% The above step is broascasting- taking vector inputs state
% variables ( x and y) and scaler (c) and choice variable(z) we get a matrix output of the
% function

f_azk = @(a,z,k) c+a+z+k;

[val, idx]=min(abs((f_azk(a_v, z_v, k))'));

rootk=k(idx);

close all;
figure();
hist(val);
figure();
hist(rootk);
```
